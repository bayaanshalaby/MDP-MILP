## 1. The problem: 

We are interested in maximizing the expected value of $y$ by controlling $x$:

$$
\arg\max_x \mathbb{E}\left[ y \middle | x \right]
$$

By the definition of expectation we get:

$$
\begin{align}
  &\arg\max_x \mathbb{E}\left[ y \middle | x \right] \\
= &\arg\max_x \mathbb{E}_{y\sim p(\cdot|x)}\left[ y\right] \\
= & \arg\max_x \int_{-\infty}^{+\infty} p(y|x) y dy \\
\end{align}
$$

Approximating the expectation using Monte-Carlo sampling:

$$
\begin{align}
  & \arg\max_x \int_{-\infty}^{+\infty} p(y|x) y dy \\
= & \arg\max_x \frac{1}{N} \sum_{i=1}^{N} \hat{y}_i
\end{align}
$$
where $\hat{y}_i \sim p(y|x)$.

## 2. Graphical models:

### Model 1: without re-parameterization

$$
\begin{align}
x & \sim \mathcal{N}(0.0, 1.0) \\
y & \sim \mathcal{N}(\min(x, 0), \max(x, 0))
\end{align}
$$

### Model 2: with re-parameterization

$$
\begin{align}
x & \sim \mathcal{N}(0.0, 1.0) \\
y & = \min(x, 0) + \max(x, 0) \cdot \epsilon \\
\epsilon & \sim \mathcal{N}(0.0, 1.0)
\end{align}
$$

### Imports

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

**if** $ x \geq 0$:

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

for x in np.logspace(-2, -1, 5):
    mu = np.minimum(x, 0)
    sigma = np.maximum(x, 0)
    
    # model 1
    y = np.random.normal(loc=mu, scale=sigma, size=100000)
    ax1.hist(y, bins=1000, density=True, histtype='step', label='x = {:.5f}'.format(x))
    ax1.set_title('WITHOUT re-parameterization')
    ax1.set_xlabel('y')
    ax1.set_ylabel('p(y|x)')
    ax1.grid()
    ax1.legend()
    
    # model 2
    epsilon = np.random.normal(size=100000)
    y = mu + sigma * epsilon
    ax2.hist(y, bins=1000, density=True, histtype='step', label='x = {:.5f}'.format(x))
    ax2.set_title('WITH re-parameterization')
    ax2.set_xlabel('y')
    ax2.set_ylabel('p(y|x)')
    ax2.grid()
    ax2.legend()
```

**if** $ x \leq 0$:

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

for x in -np.logspace(-2, -1, 5):
    mu = np.minimum(x, 0)
    sigma = np.maximum(x, 0)
    
    # model 1
    y = np.random.normal(loc=mu, scale=sigma, size=100000)
    ax1.hist(y, bins=1000, density=True, histtype='step', label='x = {:.5f}'.format(x))
    ax1.set_title('WITHOUT re-parameterization')
    ax1.set_xlabel('y')
    ax1.set_ylabel('p(y|x)')
    ax1.grid()
    ax1.legend()
    
    # model 2
    epsilon = np.random.normal(size=100000)
    y = mu + sigma * epsilon
    ax2.hist(y, bins=1000, density=True, histtype='step', label='x = {:.5f}'.format(x))
    ax2.set_title('WITH re-parameterization')
    ax2.set_xlabel('y')
    ax2.set_ylabel('p(y|x)')
    ax2.grid()
    ax2.legend()
```
