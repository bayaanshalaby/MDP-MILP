# How to Grow a Tall Tulip
### An Analysis of the Effectiveness of MILP-Based Expectation Optimization of Linear Guassian Models
Imagine that you are a gardener who is interested in maximizing the height of a tulip you plant. The height of the tulip is normally distributed with parameters that are linear or piecewise linear functions of decisions you must make prior to planting the tulip. How do you ensure that you make the best decisions to maximize the likelihood of a tall tulip?

A possible way to do this is to formulate the problem as a Mixed Integer Linear Program (MILP) by reparameterizing the objective distributions to be used in a monte-carlo'd expectation. This notebook examines the effectiveness of this method for 3 models of increasing complexity as we increase the upper bound of the monte-carlo'd summation. 

### Preliminary Definitions

Mixed Integer Linear Program (MILP): an optimization model the objective function as well as the constraints are linear functions of either continuous or integer decision variables.

Linear Gaussian Model: a one-step decision making problem where all stochasticity is Guassian with parameters that are linear or piecewise linear functions of decision variable. 

Reparameterized Gaussian Distribution: a variable y ~ N(f(x), g(x)) can be reparameterized as y = f(x) + g(x)*error, 
where error ~ N(0, 1) 

Monte Carlo'd Expectation: Approximating the expected value of a distribution by averaging a set amount of samples n from the distribution. In theory, as n approaches infinity, this approximation should approach the population mean.
